```python
import pymongo
```

```python
def get_best(db, metric, config_param=None, config_val=None, one=True):
    myclient = pymongo.MongoClient("mongodb://localhost:27017/")
    mydb = myclient[db]
    res = mydb["metrics"].aggregate([{
        "$match": {"name": metric}  # only consider metric
    },
        {"$unwind": "$values"},
        {"$group":
         {'_id': '$_id',
          'val': {'$min': "$values"}, 'run_id' : { '$first': '$run_id' }}
         },  # find min values
        {"$sort": {"val": 1}}  # sort
    ])
    
    if config_param is None:
        if one:
            return list(res)[0]
        return list(res)
    
    filtered_res = []
#   filter only results for the config
    for res_obj in res:
        run = list(mydb['runs'].find({'_id': res_obj['run_id']}))[0]
        if 'dont_emb' in run['config']:
            if run['config']['epochs']  > 0:
                res_obj['exp_id'] = run['config']['exp_id']
                filtered_res.append(res_obj)
        else:
            print('NB: just used run without considering param')
#             filtered_res.append(res_obj)
    if one:
        return filtered_res[0]
    return filtered_res
```
