```python
from pathlib import Path

import numpy as np
import pandas as pd
import swifter

import cleantext

pd.options.display.max_colwidth = 1000
```

```python
max_previous = 6

OUT = Path('~/data/ynacc_proc/replicate/threads_prev' + str(max_previous))

df_cl_tr = pd.read_csv('~/data/ynacc_proc/replicate/split/train.csv')
df_cl_va = pd.read_csv('~/data/ynacc_proc/replicate/split/val.csv')
```

```python
def create_text(row):
    the_index = row['commentindex'] + 1
    comments_before = list(row['text_threads'])[:the_index] # for this comment
    
    if not max_previous is None:
        comments_before = comments_before[-max_previous:]
    
    res = 'xx_thread_start ' + ' '.join([ 'xx_comment_start ' + xx + ' xx_comment_end' for xx in comments_before]) + ' xx_thread_end'
    assert res.endswith(row['text'] + ' xx_comment_end xx_thread_end')
    return res
```

```python
def process_df(df):
    threads = df.sort_values(by=['commentindex']).groupby('sdid').agg({'text': lambda x: list(x)}).reset_index()
    df_res = df.merge(threads, on='sdid', suffixes=['', '_threads'])
    df_res['text_threads'] = df_res.swifter.apply(create_text, axis=1)
    df_res['text_threads'] = df_res['text_threads'].swifter.apply(lambda x: cleantext.clean(x, lower=False, no_urls=True, no_emails=True, zero_digits=True))
    return df_res
```

```python
df_cl_tr = process_df(df_cl_tr)
```
