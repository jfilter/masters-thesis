```python
from pathlib import Path
import json

p = Path('/mnt/data/group07/johannes/germanlm/proc/wiki/')
```

```python
for i, fn in enumerate(p.glob('*/*')):
    with open(fn) as f:
        lines = f.readlines()
    for ii, l in enumerate(lines):
        data = json.loads(l)
        Path('/mnt/data/group07/johannes/germanlm/proc/final/wiki_' + str(i) + '_' + str(ii) + '.txt').write_text(data['text'])
```
