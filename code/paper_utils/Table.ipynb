```python
def row_start(x):
    return '\multirow{3}{*}{' + x + '} '

def split_str(text):
    rows = text.strip().split('\n')
    rows = [r.split(' & ') for r in rows]
    return rows

def reshape(li):
    out_rows = []
    for c in all_cols:
        for i, m in enumerate(['F1\\textsubscript{micro}', 'F1\\textsubscript{macro}', 'Kappa']):
            out_r = ''
            if i == 0:
                out_r += row_start(c)
            out_r += '& ' + m
            best_val = 0
            best_val_str = ''
            for l in li:
                rows = split_str(l)
                the_row = None
                for r in rows:
                    if r[0].startswith(c):
                        the_row = r
                value = the_row[i + 1][:5]
                if float(value) > best_val:
                    best_val = float(value)
                    best_val_str = value
                out_r += ' & ' + value
#             print(best_val)
#             print(best_val_str in out_r)
            out_r = out_r.replace(best_val_str, '\\textbf{' + best_val_str +'}')
            out_r += ' \\\\'
            if i == 2:
                out_r += ' \\addlinespace[2ex]'
            out_rows.append(out_r)
    print('\n'.join(out_rows))


def reshape_no_bert(li, no=[4, 9]):
    sums_micro = defaultdict(lambda: 0)
    sums_macro = defaultdict(lambda: 0)
    sums_kappa = defaultdict(lambda: 0)
    
    out_rows = []
    for c in all_cols:
        all_micros = []
        all_macros = []
        all_kappas = []
        for i, m in enumerate(['F1\\textsubscript{micro}', 'F1\\textsubscript{macro}', 'Kappa']):
            out_r = ''
            if i == 0:
                out_r += row_start(c)
            out_r += '& ' + m
            best_val = 0
            best_val_str = ''
            bert_val = 0
            bert_val_str = ''
            for l_i, l in enumerate(li):
                rows = split_str(l)
                the_row = None
                for r in rows:
                    if r[0].startswith(c):
                        the_row = r
                value = the_row[i + 1][:5]
                
                if i == 0:
                    sums_micro[l_i] += float(value)
                if i == 1:
                    sums_macro[l_i] += float(value)
                if i == 2:
                    sums_kappa[l_i] += float(value)
                
                if not l_i in no and float(value) > best_val:
                    best_val = float(value)
                    best_val_str = value
                if l_i in no and float(value) > bert_val:
                    bert_val = float(value)
                    bert_val_str = value
                    
                out_r += ' & ' + value
#             print(best_val)
#             print(best_val_str in out_r)
            out_r = out_r.replace(best_val_str, '\\textbf{' + best_val_str +'}')
            if bert_val > best_val:
                out_r = out_r.replace(bert_val_str, '\\underline{' + bert_val_str +'}')
        
            out_r += ' \\\\'
            if i == 2:
                out_r += ' \\addlinespace[2ex]'
            out_rows.append(out_r)
    final = '''\midrule
\addlinespace[2ex]
''' + row_start('Average')
    for i, m in enumerate(['F1\\textsubscript{micro}', 'F1\\textsubscript{macro}', 'Kappa']):
        final += ' & ' + m
        for l_i, _ in enumerate(li):
            if i == 0:
                final += ' & ' + str(sums_micro[l_i]/9)[:5]
            if i == 1:
                final += ' & ' + str(sums_macro[l_i]/9)[:5] 
            if i == 2:
                final += ' & ' + str(sums_kappa[l_i]/9)[:5]
        final += ' \\\\'
        if i == 2:
            final += ' \\addlinespace[2ex]'
    out_rows.append(final)
    print('\n'.join(out_rows))
        
```
