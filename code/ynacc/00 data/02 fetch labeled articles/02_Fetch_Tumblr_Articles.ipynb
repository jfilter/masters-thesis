```python
def process(bl):
    r = client.posts(bl[0], id=bl[1])
    print(bl[2])
    cr = r['posts'][0]['body']
    a = Article(bl[2])
    a.download(input_html=cr)
    a.parse()
    print(a.text)
    return a
```
