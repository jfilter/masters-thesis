```python
from concurrent.futures import ThreadPoolExecutor
from pathlib import Path

import get_wayback_machine
import pandas as pd
from newspaper import Article
```

```python
BASE_PATH = Path('/mnt/data/datasets/ydata-ynacc-v1_0')
in_path = BASE_PATH/'ydata-ynacc-v1_0_unlabeled_conversations.tsv'
```

```python
df = pd.read_csv(in_path, engine='python', sep='\t', quoting=3, error_bad_lines=False, usecols=['url'])
```

```python
urls = df['url'].unique().tolist()
```

```python
urls
```

```python
weburls = [u for u in urls if u and not '.tumblr.com' in u and u.lower() != 'nourl']
tumblrurls = [u for u in urls if u and '.tumblr.com' in u and u.lower() != 'nourl']
```

```python
len(tumblrurls)
```

```python
len(weburls)
```

```python
finals = []
with ThreadPoolExecutor(max_workers=50) as executor:
    results = executor.map(get_wayback_machine.get, weburls)
    for url, value in zip(weburls, results):
        if not value is None:
            finals.append([url, value.text])
```

```python
len(finals)
```

```python
res = []
for [url, text] in finals:
    a = Article(url)
    a.download(input_html=text)
    a.parse()
    res.append(a)
```

```python
res
```

```python
res_list = [{'url': f.url, 'text': f.text, 'publish_date': f.publish_date, 'title': f.title} for f in res]
```

```python
df = pd.DataFrame(res_list)
```
