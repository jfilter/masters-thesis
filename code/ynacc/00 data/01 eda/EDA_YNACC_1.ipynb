```python
imp_cols = ['constructiveclass', 'sd_agreement', 'sd_type', 'sentiment', 'tone', 'commentagreement', 'topic', 'intendedaudience', 'persuasiveness']
```

```python
df_a
```

```python
df_aa
```

```python
for i_c in imp_cols:
    c = alt.Chart(df_aa.reset_index()).mark_bar().encode(
        x="index",
        y=i_c,
    )
    c.display()
```
