```python
from collections import Counter
from pathlib import Path
import re

import pandas as pd
```

```python
data_folder = '/mnt/data/group07/johannes/germanlm/proc/final_merged'
paths = list(Path(data_folder).glob('*.txt'))
```

```python
chars_space = re.compile("[^a-zA-Z\s]")
one_space = re.compile('\s+')
```

```python
c = Counter()
for p in paths:
    df = pd.read_table(p, header=None)
    for _, row in df.iterrows():
#         print(row)
        text = row[0]
#         print(text)
        text = chars_space.sub('', text).strip()
        text = one_space.sub(' ', text)
#         print(text)
        le = len(one_space.findall(text))
        c[le] += 1
    print('done')
```
