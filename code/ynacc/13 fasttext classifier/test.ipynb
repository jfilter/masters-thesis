```python
from pathlib import Path
import subprocess
import re
import json
from collections import defaultdict

import news_utils.paper
```

```python


best = {}
best_macro = {}
classes = ['claudience', 'clpersuasive', 'clagreement', 'cldisagreement', 'clinformative', 'clmean', 'clcontroversial', 'cltopic', 'clsentiment']

all  = defaultdict(lambda: []) 

for cl in classes:
    for fn in Path('/mnt/data/group07/johannes/ynacc_proc/fasttext_baseline/cls/' + cl).glob('*results.json'):
#         print(fn)
        text = fn.read_text()
        all[cl].append([fn, text])

#news_utils.paper.print_table(all)
res = news_utils.paper.get_best_kappa(all)
final = []
for r in res:
    print(r)
    cl = classes.index('cl' + r[1].lower())
    print(r[0], cl)
    subprocess.run("python run_test.py --cl " + classes[cl] + ' --path ' + str(r[0]), shell=True, check=True)
    text = Path(str(r[0]) + 'test.json').read_text()
    js = json.loads(text)
    final.append(r + [js['micro avg']['f1-score'], js['macro avg']['f1-score'], js['kappa']])

for f in final:
    f = [str(x) for x in f[1:]]
    print(' & '.join(f) + ' \\\\')
```
