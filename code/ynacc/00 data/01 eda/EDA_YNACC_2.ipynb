```python
import pandas as pd
import altair as alt
alt.renderers.enable('notebook')
```

```python
PATH = '/mnt/data/datasets/ydata-ynacc-v1_0/ydata-ynacc-v1_0_turk_annotations.tsv'
```

```python
df = pd.read_table(PATH)
```

```python
imp_cols = ['constructiveclass', 'sd_agreement', 'sd_type']
```
