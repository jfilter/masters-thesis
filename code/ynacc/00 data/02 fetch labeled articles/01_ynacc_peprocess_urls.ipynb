```python
res = []
failed = []
```

```python
for u in urls[0]:
    if u in res:
        continue
    if u in failed:
        continue
        
    print(u)
    r = requests.get(f"http://archive.org/wayback/available?url={u.split('?')[0]}")
    r.raise_for_status()

    r_json = r.json()
    if not 'closest' in r_json['archived_snapshots']:
        print('what', u)
        failed.append(u)
        continue    
    clo = r_json['archived_snapshots']['closest']
    if clo['status'] != '200':
        print(clo)
        print(u)
        failed.append(u)
        continue
    rr = requests.get(clo['url'])
    
    rr.raise_for_status()
    
    a = process(u, input_html=rr.text)
    if a is None:
        print('fail')
        failed.append(u)
    else:
        res.append(a)
    
```
