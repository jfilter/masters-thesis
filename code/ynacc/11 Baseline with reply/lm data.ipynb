```python
import pandas as pd
from pathlib import Path
```

```python
OUT = Path('~/data/ynacc_proc/replicate/lmdata_complete')

BASE_PATH = Path('/mnt/data/datasets/ydata-ynacc-v1_0')
ANN1 = BASE_PATH/'ydata-ynacc-v1_0_expert_annotations.tsv'
ANN2 = BASE_PATH/'ydata-ynacc-v1_0_turk_annotations.tsv'
UNL = BASE_PATH/'ydata-ynacc-v1_0_unlabeled_conversations.tsv'
TRAIN_IDS = BASE_PATH/'ydata-ynacc-v1_0_train-ids.txt'
```

```python
!ls {BASE_PATH}
```

```python
trainids = pd.read_csv(TRAIN_IDS, header=None)
```

```python
!head {ANN1}
```

```python
df_an1 = pd.read_table(ANN1)

df_an1 = df_an1[df_an1['sdid'].isin(list(trainids[0]))]

df_an1 = df_an1.drop_duplicates(subset=['timestamp', 'text'])

df_an1
```

```python
!head {ANN2}
```

```python
df_an2 = pd.read_table(ANN2)

df_an2 = df_an2[df_an2['sdid'].isin(list(trainids[0]))]

df_an2 = df_an2.drop_duplicates(subset=['timestamp', 'text'])

df_an2
```

## ignore quoting

```python
df = pd.read_csv(UNL, engine='python', sep='\t', quoting=3, error_bad_lines=False)
```

```python
df
```

```python
df_o = pd.concat([df, df_an1, df_an2]).dropna(subset=['text']).sort_values(['timestamp'])
df_o
```

```python
df_o['text']
```

```python
df_o[:200000].to_csv(str(OUT) + '/train.csv', index=False)
```

```python
df_o[200000:].to_csv(str(OUT) + '/val.csv', index=False)
```
